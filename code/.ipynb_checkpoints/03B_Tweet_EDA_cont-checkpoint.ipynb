### Percentage of digits in User Names

```python
#Count number of digits in user name / author
def author_digits(text):
    count = 0
    for i in text:
        if i.isnumeric():
            count += 1
        else:
            count = 0
    return count / len(text)  
    
```

```python
author_digits('Hello5555')
```

```python
df['author_digits_ratio'] = df['author'].apply(author_digits)
df.head(3)
```

### Ratio of User_followers to User_Following

```python
df['user_following'].value_counts()
```

```python
df['user_followers'].value_counts()
```

```python
# Ratio of user_followers to user_following

def get_follow_ratio(followers, following):
    if followers == 0:
        return 0
    elif following == 0:
        return (followers / 1)
    else:
        return (followers / following)
```

```python
df['follow_ratio'] = [get_follow_ratio(m, n) for m, n in zip(df['user_followers'], df['user_following'])]
df.head(5)
```

### Percentage of upper characters in text

```python
# Percentage of upper case characters
def upper(text):
    upper_sum = sum(i.isupper() for i in text)
    return upper_sum / len(text)   
```

```python
upper('Hello Friend')
```

```python
df['text_upper'] = df['text'].apply(upper)
df.head(3)
```

## Linear Regression
Quickly running a model to see if the features we've explored can help model.

```python
#Create features and X, y for our model
X = df._get_numeric_data().drop(columns = ['target', 'reply_count', 'retweet_count', 'favorite_count', 'user_favorites'])

y = df['target']

# X = bikes._get_numeric_data().drop(columns=['tripduration'])

# y = bikes['tripduration']
```

```python
#Instantiate
lr = LinearRegression()

#Fit
lr.fit(X, y)
```

```python
cross_val_score(lr, X, y, cv=5).mean()  #cross val score before the split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
df['target'].mean()
```

```python
# Train score
lr.score(X_train, y_train)
```

```python
# Test score
lr.score(X_test, y_test) 
```

Unfortunately, this helps confirm what we suspected. The correlations are extremely low and the scores are as well. This is evidence that we could benefit from new data such as user profiles.
